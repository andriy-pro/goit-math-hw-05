# **Завдання 1**

Виконай диференціювання функції та знайди значення похідних у точках $x=1$ і $x=-1/2$. Розв’яжи програмно та аналітично.
1) $$\frac{x^3}{2} + \frac{x^2}{2} - 2x$$
2) $$\sqrt{x^2 + 1}$$
3) $$\frac{1}{\sqrt{x^2+1}}$$

## Розв'язання

### 1. $$\frac{x^3}{2} + \frac{x^2}{2} - 2x$$

#### Аналітично

Для знаходження похідної скористаємося наступними правилами диференціювання:

$\frac{d}{dx}(x^n) = nx^{n-1}$

$\frac{d}{dx}(cf(x)) = c \frac{d}{dx} f(x)$, де $c$ — константа.

$\frac{d}{dx}(f(x) + g(x)) = \frac{d}{dx} f(x) + \frac{d}{dx} g(x)$

Використаємо ці правила для диференціювання кожного доданка функції $\frac{x^3}{2} + \frac{x^2}{2} - 2x$:
$$
\frac{d}{dx} \left( \frac{x^3}{2} \right) = \frac{1}{2} \cdot \frac{d}{dx} (x^3) = \frac{1}{2} \cdot 3x^2 = \frac{3x^2}{2}.
$$

$$
\frac{d}{dx} \left( \frac{x^2}{2} \right) = \frac{1}{2} \cdot \frac{d}{dx} (x^2) = \frac{1}{2} \cdot 2x = x.
$$

$$
\frac{d}{dx} (-2x) = -2 \cdot \frac{d}{dx} (x) = -2 \cdot 1 = -2.
$$

Тепер складемо похідну функції:
$$
f'(x) = \frac{3x^2}{2} + x - 2.
$$

Знайдемо значення похідної в точці $x = 1$:
$$
f'(1) = \frac{3(1)^2}{2} + 1 - 2 = \frac{3}{2} - 1 = \frac{1}{2}.
$$

Знайдемо значення похідної в точці $x = -\frac{1}{2}$:
$$
\begin{gather}
f'\left(-\frac{1}{2}\right) = \frac{3\left(-\frac{1}{2}\right)^2}{2} + \left(-\frac{1}{2}\right) - 2
= \frac{3 \cdot \frac{1}{4}}{2} - \frac{1}{2} - 2 \\
= \frac{3}{8} - \frac{4}{8} - \frac{16}{8} 
= \frac{3 - 4 - 16}{8}
= \frac{-17}{8}
\end{gather}
$$


Отже, значення похідної в заданих точках:
$$
f'(1) = \frac{1}{2},
$$
$$
f'\left(-\frac{1}{2}\right) = -2\frac{1}{8}.
$$


#### Програмно

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Визначаємо змінну
x = sp.symbols('x')
point_1 = 1
point_2 = -1/2

# Визначаємо функцію
f = (x**3)/2 + (x**2)/2 - 2*x

# Знаходимо похідну функції
derivative = sp.diff(f, x)

# Обчислюємо значення похідної в точці x = 1
derivative_point_1 = derivative.subs(x, point_1)

# Обчислюємо значення похідної в точці x = -1/2
derivative_point_2 = derivative.subs(x, point_2)

# Виводимо результати
print("Функція:")
display(f)
print("Похідна:")
display(derivative)
print(f"Похідна в точці x=1: f'(1) = {derivative_point_1}")
print(f"Похідна в точці x=-1/2: f'(-1/2) = {round(derivative_point_2, 3)}")

# Перетворюємо функції на lambdify для обчислень
f_lambdified = sp.lambdify(x, f, "numpy")
derivative_lambdified = sp.lambdify(x, derivative, "numpy")

# Генеруємо значення для графіку
x_vals = np.linspace(-2, 2, 400)
f_vals = f_lambdified(x_vals)
derivative_vals = derivative_lambdified(x_vals)

# Створюємо графік
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='$f(x)$', color='blue')
plt.plot(x_vals, derivative_vals, label="$f'(x)$", color='orange')
plt.scatter([point_1, point_2], [derivative_point_1, derivative_point_2], color='green')
plt.text(point_1, derivative_point_1, f"({point_1}, {derivative_point_1})", fontsize=12, ha='right')
plt.text(point_2, derivative_point_2, f"({round(point_2, 3)}, {round(derivative_point_2, 3)})", fontsize=12, ha='left')

# Налаштування графіку
plt.title('Графік функції $f(x)$ та її похідної $f\'(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
